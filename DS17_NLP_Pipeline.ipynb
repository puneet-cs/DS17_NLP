# 2. Lowercasing

```python
text = "SMIT : HOW ARE YOU. CRUSH : GO TO HELL."

# Use SpaCy for lowercasing

doc = nlp(text)

for token in doc:
    print(token.lower_)
```

```python
# Sentence-wise lowercasing using spaCy 
for token in doc.sents:
    print(token.text.lower())
```

```python
# word-wise lowercasing using nltk

tokens = word_tokenize(text)

for t in tokens:
    print(t.lower())
```

```python
# sentencewise lowercasing using nltk
sentences = sent_tokenize(text)

for sent in sentences:
    print(sent.lower())
```

```python
# BERT for lowercasing

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

tokens = tokenizer.tokenize(text)

print(tokens)
```

```python
# lowercasing using python

lowercase_text = text.lower()

print(lowercase_text)
```

# 3. Stopword removal

```python
# stopword removal using spaCy

text = "NLP is important skill. This may land you in high paying job."

doc = nlp(text)

for token in doc:
    if token.is_stop:
        print(token.text)
```

```python
# stopword using nltk

from nltk.corpus import stopwords

tokens = word_tokenize(text)

stop_words = set(stopwords.words("english"))

print(stop_words)
```

```python
for word in tokens:
    if word.lower() not in stop_words:
        print(word)
```

```python
# stopword using transformer

tokens = tokenizer.tokenize(text)

stop_words = set(stopwords.words("english"))

for word in tokens:
    if word in stop_words:
        print(word)
```

# Stemming & Lemmatization

```python
# Stemming

text = "running studying unhappy sleeping better studies"

# using nltk for stemming
from nltk.stem import PorterStemmer

ps = PorterStemmer()

tokens = word_tokenize(text)

for word in tokens:
    print(ps.stem(word))
```

```python
# improved version of porter stemmer
from nltk.stem import SnowballStemmer
sb = SnowballStemmer("english")
tokens = word_tokenize(text)

for word in tokens:
    print(sb.stem(word))
```

```python
# Lemmatization using spaCy

doc = nlp(text)

for token in doc:
    print(token.lemma_)
```

```python
# Lemmatization using nltk
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

tokens = word_tokenize(text)

for word in tokens:
    print(wnl.lemmatize(word))
```

# Removing Punchuation and Numbers

```python
text = "Chaitanya have a GF! She spend everyday $1000 on shopping? They are in relation from 2020"

# Using SpaCy

doc = nlp(text)

for token in doc:
    if not token.is_punct and not token.like_num:
        print(token)
```

```python
# nltk + Python (String)

import string

tokens = word_tokenize(text)

for word in tokens:
    if word not in string.punctuation and not word.isdigit():
        print(word)
```

```python
# transformers + string
tokens = tokenizer.tokenize(text)

for word in tokens:
    if word not in string.punctuation and not word.isdigit():
        print(word)
```

# Linguistic Analysis / Feature extraction

# N-gram

```python
# N-gram using SpaCy

text = "Varun GF checks his phone daily, She is Suspesious"

doc = nlp(text)

for i in range(len(doc) - 1):
    bigram = doc[i].text + " " + doc[i+1].text
    print(bigram)
```

```python
# N-gram using nltk

from nltk.util import ngrams

tokens = word_tokenize(text)

trigram = list(ngrams(tokens, 5))
print(trigram)
```

```python
# ngram using transformer

tokens = tokenizer.tokenize(text)

for i in range(len(tokens) - 1):
    bigram =  tokens[i] + " " + tokens[i+1]
    print(bigram)
```

# Part of Speech

```python
# POS usign SpaCy

text = "Visnu girlfirend daily comes in his dream and make him afraid"

doc = nlp(text)

for token in doc:
    print(token.text, token.pos_)
```

```python
# pos using nltk

tokens = word_tokenize(text)

pos = nltk.pos_tag(tokens)

print(pos)
```

```python
# pos using transformer

from transformers import pipeline

pos = pipeline("token-classification", model = "vblagoje/bert-english-uncased-finetuned-pos")

result = pos(text)

for w in result:
    print(w['word'], w['entity'])
```

# Named Entity Recognition (NER)

```python
# NER using SpaCy

text = "Taj and Jaffer went for wonderla on 13/03/2026 in Hyderabad"

doc = nlp(text)

for token in doc.ents:
    print(token.text, token.label_)
```

```python
# NER using nltk

from nltk import ne_chunk

tokens = word_tokenize(text)

pos = nltk.pos_tag(tokens)

ner = ne_chunk(pos )

print(ner)
```

```python
# NER using transformer

ner = pipeline("ner", model = "dslim/bert-base-NER")

result = ner(text)

for r in result:
    print(r["word"], r["entity"])
```

# Text Representation

### Bag of Words

```python
# BoW using SpaCy + Python library

from collections import Counter

text = "Kaaki kootha koosindi, kokila kootha koosindi,kaaki kootha kokila kootha kante konchem kathinam"

doc = nlp(text)

word_freq = []

for token in doc:
    word_freq.append(token.text.lower())

bow = Counter(word_freq)

print(bow)
```

```python
# BOW using nltk + Counter

tokens = word_tokenize(text.lower())

bow = Counter(tokens)

print(bow)
```

```python
# BOW using transformer + Counter

tokens = tokenizer.tokenize(text.lower())

bow = Counter(tokens)

print(bow)
```

# TF-IDF

```python
# using Scikit-learn

from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "ranu Bombai ki ranu",  
    "Naa Ranu cheli ranu"
]

tfidf = TfidfVectorizer()

X = tfidf.fit_transform(documents)

# print the important VOCAB from all the documents
print(tfidf.get_feature_names_out())

# importance of learned VOCAB for every document point of view
print(X.toarray())
```
