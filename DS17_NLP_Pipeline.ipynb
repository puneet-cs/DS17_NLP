```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# Word Tokenization using spacy

doc = nlp("Vinnu GF runs away with his Best friend. He will kill him")

for token in doc:
    print(token.text)
```

```python
# Sentence Tokenization using spacy

for sent in doc.sents:
    print(sent.text)
```

# Tokenization using NlTK

```python
import nltk
```

```python
nltk.download('all')
```

```python
# word tokenization using nltk

from nltk.tokenize import word_tokenize

text = "Varun friend got some mental issue after having 100 affairs. But still he is very Happy"

tokens = word_tokenize(text)

print(tokens)
```

```python
# Sentence tokenization using nltk

from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)

print(sentences)
```

# 2. Lowercasing

```python
text = "SMIT : HOW ARE YOU. CRUSH : GO TO HELL."

# Use SpaCy for lowercasing

doc = nlp(text)

for token in doc:
    print(token.lower_)
```

```python
# Sentence-wise lowercasing using spaCy 
for token in doc.sents:
    print(token.text.lower())
```

```python
# word-wise lowercasing using nltk

tokens = word_tokenize(text)

for t in tokens:
    print(t.lower())
```

```python
# sentencewise lowercasing using nltk
sentences = sent_tokenize(text)

for sent in sentences:
    print(sent.lower())
```

```python
# BERT for lowercasing

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

tokens = tokenizer.tokenize(text)

print(tokens)
```

```python
# lowercasing using python

lowercase_text = text.lower()

print(lowercase_text)
```

# 3. Stopword removal

```python
# stopword removal using spaCy

text = "NLP is important skill. This may land you in high paying job."

doc = nlp(text)

for token in doc:
    if token.is_stop:
        print(token.text)
```

```python
# stopword using nltk

from nltk.corpus import stopwords

tokens = word_tokenize(text)

stop_words = set(stopwords.words("english"))

print(stop_words)
```

```python
for word in tokens:
    if word.lower() not in stop_words:
        print(word)
```

```python
# stopword using transformer

tokens = tokenizer.tokenize(text)

stop_words = set(stopwords.words("english"))

for word in tokens:
    if word in stop_words:
        print(word)
```
